## Exploratory Data Analysis


We'll be importing data from the [Freddie Mac Database](http://www.freddiemac.com/research/datasets/sf_loanlevel_dataset.page) as a pandas dataframe.  We will need 

```python
# download_url = "https://raw.githubusercontent.com/annahaensch/DataAndSocialJustice/main/Data/Freddie_Mac/historical_data_2019Q1.txt"

url = "historical_data_2019Q1.txt"

file = open(url, 'r')
lines = file.readlines()
```

```python
# Initialize empty dataframe
df = pd.DataFrame(columns = [i for i in range(31)])

# Add rows line by line
for line in lines[:1000:]: # Take only first 1000 lines to avoid overflow error. 
    new_df_row = [s.strip("\n") for s in line.split("|")]
    df.loc[len(df)] = new_df_row
    
df.head()
```
